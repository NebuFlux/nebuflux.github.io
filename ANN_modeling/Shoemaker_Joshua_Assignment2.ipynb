```python
NB_EPOCH = 20
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10 # number of outputs = number of digits
OPTIMIZER = SGD() # SGD optimizer, explained later in this chapter
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2 # how much TRAIN is reserved for VALIDATION
```

### Spliting Data Into Train and Test Sets

```python
# data: shuffled and split between train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

---
### Shaping

```python
# X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784

RESHAPED = 784

X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
# normalize
X_train /= 255
X_test /= 255

print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
# convert class vectors to binary class matrices

# Y_train = np_utils.to_categorical(y_train, NB_CLASSES) # From the book, code replaced below
Y_train = to_categorical(y_train, NB_CLASSES) # Updated for current version

#Y_test = np_utils.to_categorical(y_test, NB_CLASSES) # From the book, code replaced below
Y_test = to_categorical(y_test, NB_CLASSES) # Updated for current version
```

## Accuracy Rates and Parameter Changes

### Base Case Two Hidden layers
In the base case, we use the book's standard parameters for two hidden layers with 128 neurons each. This results in a training accuracy of 94.67%, validation accuracy of 94.59%, and test accuracy of 94.30%.

### Experiment 1: Learning Rate
The default Stochastic Gradient Descent uses a constant learning rate to move weights during the learning phase of back propagation. As a default, the value is set to 0.01. Experiment 1 changes the learning rate to 0.001, a tenth of the default. This means the gradient moves the weights in smaller increments towards 0 or 1, effectively slowing the learning progress. The result is a training accuracy of 87.34%, validation accuracy 88.47%, and test accuracy 88.38% after 20 weight updates.

### Experiment 2: Optimizer
The second experiment introduces the Adam optimizer, resulting in a training accuracy of 99.58%, validation accuracy of 97.50%, and test accuracy of 97.71%. Adam provides a massive increase in accuracy across all metrics. This is due to the fundamental differences between the default stochastic gradient descent and the adaptive moment estimation algorithm. Adam uses an adaptive learning with 1st order and 2nd order moments ([Keras, 2025](https://keras.io/api/optimizers/adam/)). Essentially, Adam can change the magnitude at which weights change relative to the loss gradient, or how wrong the weight was. This is also called root-mean-square propagation. Adam also uses momentum, which adds a fraction of the previous gradient to the current gradient used to calculate the magnitude of change for the weight ([Rahul Agarwal](https://builtin.com/machine-learning/adam-optimization)). Together, the 1st order and 2nd order moments work to produce highly efficient weight modification, resulting in quicker learning and lower epochs.


### Experiment 3: Maxing Out
Experiment three seeks to reach the highest possible test accuracy scores in the model by utilizing an efficient optimizer, quadrupling the epochs, and doubling the neurons in the two hidden layers. This results in a final training accuracy of 100%, validation accuracy of 98.34%, and test accuracy of 98.29%. This results in less than a percent increase over experiment 2, which only changes the optimizer from the base case. Reviewing the book's graphs on pages 32-37 shows that experiment 3 has essentially reached the highest accuracy this simple model can achieve by varying its basic parameters.
